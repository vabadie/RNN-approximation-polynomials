## Approximation of the square function
The program below allows to plot the output of the squaring RNN (Theorem 8) after a selected amount of iterations.

```python
# ====== Parameters of the experiment ======
full_precision_square = True      # Set to True if you want to use high-precision arithmetic but potentially (very) long runtimes. The precision required is automatically determined by the parameters of the experiment. Set to False if you want to use standard double-precision floating point arithmetic with much faster runtimes but less precision. Note that the error will then be limited to around 1e-30 due to numerical precision issues.
n_points_square = 100              # number of points to plot the target function and the RNN approximation, and to evaluate the error
T_square = 20                     # number of iterations to run the RNN for
t_values_square = [1, 2, 5, 10]    # list of time indices to plot
D_square = 1.0                     # to control the size [-D,D] of the segment on which we measure the error


from rnn_polynomial_approximation_support import run_square_approximation_experiment
globals().update(run_square_approximation_experiment(
    full_precision=full_precision_square,
    mp_dps=2 * T_square + 50,  # set the precision for mpmath to be high enough to handle the precision required by the experiment (this is automatically determined based on the number of iterations T, but you can adjust it if you want)
    n_points=n_points_square,
    x_min=-D_square,
    x_max=D_square,
    T=T_square,
    t_values=t_values_square,
    D_square=D_square,
    plot_target_curve=True,
    plot_error_across_iterations=True,
    eps_plot=1e-60 if full_precision_square else 1e-30,
))
```
